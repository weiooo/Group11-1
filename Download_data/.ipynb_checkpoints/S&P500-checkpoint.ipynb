```python
import pandas as pd
import requests as re
from bs4 import BeautifulSoup

soup = BeautifulSoup(re.get("http://slickcharts.com/sp500").text, "html5lib")
content = soup.find_all('table')[0].find_all('tr')[1:]

name_list = []
label_list = []
weight_list = []

for i in content:
    name_list.append(i.find_all('td')[1].get_text())
    label_list.append(i.find_all('td')[2].get_text().replace('.','-'))
    weight_list.append(float(i.find_all('td')[3].get_text()))
    
df = pd.DataFrame({'name': name_list, 'label': label_list, 'weight': weight_list})
stk_list = df.label
print(stk_list)
```

```python
# Adjust_close price: use yf.download
import yfinance as yf
import time
import datetime 

start = datetime.date(2009,1,2)
end = datetime.date(2019,1,1)

for i in stk_list:
    
    print('processing: ' + i)
    # to csv
    yf.download(i,start,end).to_csv('price_'+i+'.csv')
    time.sleep(1)
```

```python
# NoAdjust_close price: use yf.Ticker
import yfinance as yf
import time

# S&P500
for i in stk_list:
    
    # progressing
    print('processing: ' + i)
    # to csv
    stock = yf.Ticker(i)
    stock.history(start="2009-12-02", end="2019-12-01").to_csv('price_'+i+'.csv')
    time.sleep(1)
```
